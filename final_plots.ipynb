```python
import numpy as np
from matplotlib import pyplot as plt
from scipy import io
import os.path as osp
import matplotlib
matplotlib.rc('font', size=12)
```

```python
# The values are observed by running mgda or mgda_ub
loss_mgda_ub = [0.22587038119057062, 0.22587038119057062, 0.22587038119057062]
loss_mgda = [0.08184345161776997, 0.08184345161776997, 0.08184345161776997]
loss_mgda = [l/256 for l in loss_mgda]
loss_mgda_ub = [l/256 for l in loss_mgda_ub]
```

```python
dat = np.load("mtl_scalarization.npy")
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(projection='3d')
ax.scatter(dat[:,0], dat[:,1], dat[:,2], label="Scalarization")
ax.scatter(loss_mgda_ub[0], loss_mgda_ub[1], loss_mgda_ub[2], c='r', marker='.', s=100, label="MGDA UB")
ax.scatter(loss_mgda[0], loss_mgda[1], loss_mgda[2], c='g', marker='.', s=100, label="MGDA")

ax.set_xlabel(r'$MSE_1$')
ax.set_ylabel(r'$MSE_2$')
ax.set_zlabel(r'$MSE_3$')
# ax.invert_zaxis()
# ax.invert_yaxis()

plt.legend()

plt.show()
```
